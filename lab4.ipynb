Generators

Exercise 1:

```python
def myfunc(a):
    for i in range(1, a):
        yield i**2
a=int(input())
for nums in myfunc(a):
    print(nums)
```

Exercise 2:

```python
def myfunc(n):
    for i in range(0, n):
        if i%2==0:
            yield i
n=int(input())
for nums in myfunc(n):
    print(nums, end=",")
```

Exercise 3:

```python
def myfunc(n):
    for i in range(0, n):
        if i%3==0 and i%4==0:
            yield i
n=int(input())
for nums in myfunc(n):
    print(nums)
```

Exercise 4:

```python
def squares(a, b):
    for i in range(a, b):
        yield i**2
a=int(input())
b=int(input())
for square in squares(a, b):
    print(square)
```

Exercise 5: 

```python
def myfunc(a):
    for i in range(a, 0, -1):
        yield i
a=int(input())
for nums in myfunc(a):
    print(nums)
```
